```python
Metrics =covid_data.loc[:, ["state","death","recovered","positive","totalTestResults"]]
Metrics
```

```python
#positive ratio from total test results(PR)
Metrics["positive rate"]=Metrics['positive']/Metrics['totalTestResults']

```

```python
#what states had the highest posiive rate up to Mar 2021!
positive_rate=Metrics.sort_values("positive rate", ascending = False)
positive_rate
```

```python
#death ratio from positive results
Metrics["death rate"]=Metrics['death']/Metrics['positive']
```

```python
#what states had the highest deaths up to Mar 2021!
Metrics.sort_values("death", ascending = False)
```

```python
#what states had the highest death rate up to Mar 2021!
Metrics.sort_values("death rate", ascending = False)
```

```python
import pandas as pd
import numpy as np
region_data= pd.read_csv("regions.csv")
region_data
```

```python
joined_data=metrics_ethnic_data.merge(region_data)
joined_data
```

```python
#duplicate check
joined_data.shape
```

```python
regionsgrouped=joined_data.groupby(['Region']).mean()
regionsgrouped
```

```python
#In what region is the death rate highest?
regionsgrouped.sort_values("death rate", ascending = False)
```

```python
#In what region is the positive rate highest?
regionsgrouped.sort_values("positive rate", ascending = False)
```

```python
regions_stategrouped=joined_data.groupby(['Region','state']).mean()
regions_stategrouped
```

```python
#Conditional formatting

regions_stategrouped['death detail']=regions_stategrouped.apply(lambda x: 'high' if x['death rate'] > 0.02 
  else 'lower', axis =1)
regions_stategrouped                                                            
```

```python
regions_stategrouped[regions_stategrouped['death detail']=='high']
```
